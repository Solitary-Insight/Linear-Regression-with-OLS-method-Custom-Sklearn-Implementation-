```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
data=pd.read_csv("../../placement.csv")
```

```python
data.describe()
```

```python
data.info
```

```python
plt.scatter(data["cgpa"],data["package"],)
```

```python
X=data['cgpa']
```

```python
Y=data["package"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.8)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train.values.reshape(-1,1),y_train)
```

```python
predicted=lr.predict(x_test.values.reshape(-1,1))
```

```python
print("MAE",mean_absolute_error(y_test,predicted))
print("MSE",mean_squared_error(y_test,predicted))
print("F1-Score",r2_score(y_test,predicted))
```

```python
plt.scatter(x_test,y_test)
y_line=lr.intercept_+lr.coef_*x_test
plt.plot(x_test,y_line,color='red')
```

```python
plt.scatter(x_train,y_train)
y_line=lr.intercept_+lr.coef_*x_train
plt.plot(x_train,y_line,color='red')
```

```python
plt.scatter(X,Y)
y_line=lr.intercept_+lr.coef_*X
plt.plot(X,y_line,color='red')
```
